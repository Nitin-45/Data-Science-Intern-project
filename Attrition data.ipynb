# What is meant by employee attrition?
What is employee attrition?
Employee attrition is the gradual reduction in employee numbers.
Employee attrition happens when the size of your workforce diminishes over time. 
This means that employees are leaving faster than they are hired.

# Exploratory Data Analysis

# Now We Start The Visualization Part

#  We are find out wich factors affecting attrition

```python
sns.countplot(x='Attrition',data=df,hue='Gender')
#sns.set(rc={'figure.figsize':(5,5)})
plt.xlabel('Attrition',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Attrition By Gender',fontsize=25)
plt.show
```

```python
sns.countplot(hue='Attrition',y='Department',data=df,palette='CMRmap')
plt.ylabel('Department',fontsize=20)
plt.xlabel('count',fontsize=20)
plt.title('Attrition By Department',fontsize=25)
plt.show
```

```python
sns.countplot(x='PerformanceRating',hue='Attrition',data=df)
plt.xlabel('Performance Rating',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Attrition By Performance Rating',fontsize=25)
plt.show
```

```python
sns.histplot(hue='Attrition',data=df,y='EducationField',palette='cividis')
plt.ylabel('Education Field',fontsize=20)
plt.xlabel('count',fontsize=20)
plt.title('Attrition By Education Field',fontsize=25)
plt.show
```

```python
sns.boxplot(hue='Attrition',y='TotalWorkingYears',data=df,width=0.5,x='Department',palette='CMRmap')
plt.ylabel('Total Working Years',fontsize=20)
plt.xlabel('Department',fontsize=20)
plt.title('Attrition By Total Working Years',fontsize=25)
plt.show
```

```python
sns.histplot(hue='Attrition',x='Age',data=df)
plt.xlabel('Age',fontsize=20)
plt.ylabel('count',fontsize=20)
plt.title('Attrition By Age',fontsize=25)
plt.show
```

```python
sns.countplot(hue='Attrition',x='YearsSinceLastPromotion',data=df)
plt.xlabel('Years Since Last Promotion',fontsize=20)
plt.ylabel('count',fontsize=20)
plt.title('Attrition By Years Since Last Promotion',fontsize=25)
plt.show
```

```python
sns.countplot(hue='Attrition',x='JobSatisfaction',data=df,palette='cividis')
plt.xlabel('Job Satisfaction',fontsize=20)
plt.ylabel('count',fontsize=20)
plt.title('Attrition By Job Satisfaction',fontsize=25)
plt.show
```

# Exploratory Data Analysis

# Most of The Order Place Year

```python
sns.histplot(x='Order Date',data=df,kde=True,color='orange')
```

# Online and offline Total Revenue

```python
sns.histplot(x='Total Revenue',data=df,hue='Sales Channel',element='poly')
plt.grid()
```

# Total Units Sold By Region 

```python
sns.barplot(y='Region',x='Units Sold',data=df,palette='CMRmap')
```

# Total Profit By Item Type

```python
sns.displot(x='Total Profit',hue='Item Type',data=df)
```

# Total Profit by Order Priority

```python
sns.boxplot(y='Total Profit',data=df,x='Order Priority',color='pink',linewidth=0.5)
```

# Unit Cost By Item Type

```python
sns.barplot(x='Unit Cost',y='Item Type',data=df,palette='CMRmap')
```

# Order Year By Region

```python
sns.histplot(y='Order Date',hue='Region',data=df)
```

# Year on Year Growth 2010 to 2017

```python
sns.lineplot(x='Order Date',y='Total Profit',data=df,color='black')
#sns.lineplot(x='Order Date',y='Total Profit',data=df,color='r')
sns.set(rc={'figure.figsize':(15,5)})
plt.xlabel('Order Year',fontsize=25)
plt.ylabel('Total Profit',fontsize=25)
plt.title('Year on Year Growth',fontsize=30)
```

# Total Profit By Region

```python
sns.barplot(y='Total Profit',x='Region',data=df,palette='cividis',errorbar=None)
```
